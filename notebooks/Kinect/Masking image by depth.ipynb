```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('depth.jpg', cv2.IMREAD_GRAYSCALE) 
plt.imshow(img, cmap='gray')
```

```python
cv2.namedWindow("Depth")
cv2.createTrackbar("d", "Depth", 0, 255, nothing)
while True:
    d = cv2.getTrackbarPos("d", "Depth")
    # TODO: optimize masking function 
    (T, threshInv) = cv2.threshold(img, d, 255,cv2.THRESH_BINARY_INV)
#     masked = np.array(list(map(lambda x: x if x < d else 255, img.flatten()))).reshape((480, 640)).astype('uint8')
    cv2.imshow("Depth", threshInv)
    key = cv2.waitKey(1)
    if key == 27:
        break
cv2.destroyAllWindows()
```
