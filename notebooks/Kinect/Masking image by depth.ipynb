# Read images

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

depth = cv2.imread('depth.jpg', cv2.IMREAD_GRAYSCALE) 
depth_with_pretty = cv2.imread('depth_with_pretty.jpg', cv2.IMREAD_GRAYSCALE) 
rgbImage = cv2.imread('rgb.jpg', cv2.IMREAD_COLOR)

plt.imshow(depth, cmap='gray')
```

```python
plt.imshow(depth_with_pretty, cmap='gray')
```

```python
plt.imshow(rgbImage)
```

# Blur images

```python
blured_depth = cv2.blur(depth, (5, 5))
plt.imshow(blured_depth, cmap='gray')
```

# Masking rgb image

```python
def nothing(x):
    pass
```

```python
cv2.namedWindow("Depth")
cv2.namedWindow("Masked")

cv2.createTrackbar("d", "Depth", 0, 255, nothing)

while True:
    d = cv2.getTrackbarPos("d", "Depth")
    _, threshInv = cv2.threshold(depth, d, 255,cv2.THRESH_BINARY_INV)
    masked_image = cv2.bitwise_and(rgbImage, rgbImage, mask=threshInv)
    cv2.imshow("Depth", threshInv)
    cv2.imshow("Masked", masked_image)
    key = cv2.waitKey(1)
    if key == 27:
        break
cv2.destroyAllWindows()
```
